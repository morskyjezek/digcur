```python
import requests

searchEndpoint = 'http://www.loc.gov/search/'
```

```python
parameters = dict()

parameters = {
    'fo' : 'json',
    'q'  : 'kittens',
    'fa' : 'online-format:image'
}

print(parameters)
```

```python
r = requests.get(searchEndpoint, parameters)

print('You requested',r.url)
print('HTTP request response code:',r)
print('Headers:',r.headers)
print('The content type of the response:',r.headers['Content-Type'])
```

```python
content = r.text
```

```python
print(content)
```

```python
type(content)
```

```python
import json
```

```python
responseJson = json.loads(content)

print(responseJson)
```

```python
#query the data some more
for key in responseJson:
    print(key)
```

```python
fname = 'kitten-result-'
format = '.json'
n = 0

for item in responseJson['results']:
    n = n + 1
    file = fname+str(n)+format
    #print(file)
    with open(file, 'w') as f:
        f.write(json.dumps(item))
        print('Wrote file',file)
print('Wrote the files!')
```
